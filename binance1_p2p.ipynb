```python
import json
import requests
import pandas as pd
import numpy as np
import time
from matplotlib import pyplot as plt
import warnings
from datetime import datetime, timedelta
import os
warnings.filterwarnings('ignore')

#Clearing old files
directory = os.getcwd()
try:
  os.remove("historic_p2p.xlsx")
except:
  print("No excel file found... No delete made... ")


def runData(fiat,type,payTypes):
    return  {
    "fiat": fiat,
    "page": 1,
    "rows": 10,
    "tradeType": type,
    "asset": "USDT",
    "countries": [],
    "proMerchantAds": False, #false. Original value
    "shieldMerchantAds": False,  #false. Original value
    "publisherType": None, #null. Original value
    "payTypes": payTypes

}

#while True: Si pongo esto se actualizara de manera constante

columns = {'Side': [], 'Nickname': [],'Minimum': [], 'Maximum': [],'Balance': [], 'Rating': [],'Completion': [],'Price':[]}
df = pd.DataFrame(data=columns)
orderType = ["BUY","SELL"]

def runMain():

    for order in orderType:

        result = requests.post("https://p2p.binance.com/bapi/c2c/v2/friendly/c2c/adv/search",json=runData("USD",order,["Zinli"]))
        result = json.loads(result.text)




        for i in range(10):
            userName = result["data"][i]["advertiser"]["nickName"]
            minQuant = result["data"][i]["adv"]["minSingleTransQuantity"]
            maxQuant = result["data"][i]["adv"]["dynamicMaxSingleTransAmount"]
            balance = result["data"][i]["adv"]["tradableQuantity"]
            rating = float(result["data"][i]["advertiser"]["positiveRate"])*100
            completion = float(result["data"][i]["advertiser"]["monthFinishRate"])*100
            price = float(result["data"][i]["adv"]["price"])
            df.loc[len(df.index)] = [order,userName, minQuant, maxQuant,balance,rating,completion,price]

    #Separate DFbuy and DFSell


historic = pd.DataFrame({"avgDiff":[],"avgSellPrice":[],"avgBuyPrice":[],"t":[]})

starttime = time.time()
while True:
    runMain()
    dfBuy = df[(df.Side == "BUY")]
    dfSell = df[(df.Side == "SELL")]
    print(dfBuy)
    print(dfSell)

    diffrence = ( dfBuy.Price.mean()-dfSell.Price.mean() ) *100
    historic.loc[len(historic.index)] = [diffrence,dfSell.Price.mean(),dfBuy.Price.mean(),datetime.now()-timedelta(hours=5)]

    print(datetime.now()-timedelta(hours=5))
    historic.to_excel("historic_p2p.xlsx")

    fig, axs = plt.subplots(2, 2)
    axs[0, 0].plot(historic.t, historic.avgDiff, 'tab:orange')
    axs[0, 0].set_title('AvgDiff')
    axs[0, 1].plot(historic.t, historic.avgSellPrice, 'tab:green')
    axs[0, 1].set_title('avgSell')
    axs[1, 0].plot(historic.t, historic.avgBuyPrice, 'tab:red')
    axs[1, 0].set_title('avgBuy')

    plt.show()

    df = pd.DataFrame(columns=df.columns)
    dfBuy = pd.DataFrame(columns=dfBuy.columns)
    dfSell = pd.DataFrame(columns=dfSell.columns)
    time.sleep(120.0 - ((time.time() - starttime) % 120.0))
```
